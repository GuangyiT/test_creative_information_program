```python
# (3)
class Enemy(object):
    def __init__(self):
        self.x = int(4)
        self.y = int(0)
        self.mark = 'O'
        self.vector = [1, 1]
    
    def __repr__(self):
        return 'x: {0}, y: {1}, vec: {2}'.format(self.x, self.y, self.vector)
    
     def move(self):
        self.x += self.vector[0]
        self.y += self.vector[1]
        if (self.y > 14):
            self.x = int(4)
            self.y = int(0)
            self.vector = [1, 1]
        elif (self.x == 0):
            self.vector[0] = 1
        elif (self.x == 8):
            self.vector[0] = -1     

class Bullet(object):
    def __init__(self, x:int):
        self.x = int(x)
        self.y = int(14)
        self.mark = 'e'
    def __repr__(self):
        return 'x: {0}, y: {1}'.format(self.x, self.y)
    def move(self):
        self.y -= 1  

class Gun(object):
    def __init__(self):
        self.x = int(4)
        self.mark = 'X'
        # 残弾数
        self.remain_bullet = 2
    def __repr__(self):
        return 'x: {0}, 残弾数: {1}'.format(self.x, self.remain_bullet)

class Game(object):
    def __init__(self):
        self.board = [
            ['|', '-', '-', '-', 'V', '-', '-', '-', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', '-', '-', '-', 'X', '-', '-', '-', '|'],            
        ]
        self.enemy = Enemy()
        self.bullets = []
        self.gun = Gun()
        self.point = 0
        self.fail = 0
        # 初期段階
        self.phase = 0
        
    def print_board(self):
        row = len(self.board)        
        print()
        col = len(self.board[0])
        for i in range(0, row):
            row_str = ''
            for j in range(0, col):
                row_str += self.board[i][j]
            print(row_str) 
    
    def fire_gun(self, key):
        if (self.gun.remain_bullet > 0 and key == 'i'):
            self.bullets.append(Bullet(self.gun.x))
            self.gun.remain_bullet -= 1           
    
    def update_enemy_board(self):
        # 元の位置を更新
        if (self.enemy.x == 4 and self.enemy.y == 0):
            self.board[0][4] = 'V'
        elif (self.enemy.x == self.gun.x and self.enemy.y == 14):
            self.board[14][self.gun.x] = 'X'
        elif (self.enemy.x == 0 or self.enemy.x == 8):
            self.board[self.enemy.y][self.enemy.x] = '|'
        else:
            self.board[self.enemy.y][self.enemy.x] = ' '
        # fail
        if (self.enemy.y == 14):
            self.fail += 1
            # reset gun bullet
            self.gun.remain_bullet = 2
        # 移動
        self.enemy.move()
        # 移動後を更新
        self.board[self.enemy.y][self.enemy.x] = 'O'
    
    def update_bullets_board(self):
        for index, bullet in enumerate(self.bullets):
            if (bullet.y >= 0):
                # 元の位置のマークが次のenemyの位置かどうか
                is_enemy = (self.board[bullet.y][bullet.x] == 'O')
                if (not is_enemy):
                    if (bullet.x == 4 and bullet.y == 0):
                        self.board[0][4] = 'V'
                    elif (bullet.x == self.gun.x and bullet.y == 14):
                        self.board[14][self.gun.x] = 'X'
                    elif (bullet.x == 0 or bullet.x == 8):
                        self.board[bullet.y][self.bullet.x] = '|'
                    else:
                        self.board[bullet.y][bullet.x] = ' '
                # 移動
                bullet.move()
                # 移動後を更新
                if (bullet.y >= 0):
                    # 移動後がenemyと当たった場合
                    if (self.board[bullet.y][bullet.x] == 'O'):                        
                        if (bullet.x == 4 and bullet.y == 0):
                            self.board[0][4] = 'V'
                        elif (bullet.x == self.gun.x and bullet.y == 14):
                            self.board[14][self.gun.x] = 'X'
                        elif (bullet.x == 0 or bullet.x == 8):
                            self.board[bullet.y][bullet.x] = '|'
                        else:
                            self.board[bullet.y][bullet.x] = ' '
                        bullet.y = -1
                        self.point += 1
                        self.enemy = Enemy()
                        self.gun.remain_bullet = 2    
                    else:    
                        self.board[bullet.y][bullet.x] = 'e'  
                    
    def update_phase(self, key):
        self.phase += 1
        self.fire_gun(key)
        self.update_enemy_board()
        self.update_bullets_board()
```

```python
# (5)
import random    

class Enemy(object):
    def __init__(self):
        self.x = int(4)
        self.y = int(0)
        self.mark = 'O'
        self.vector = [1, 1]
    
    def __repr__(self):
        return 'x: {0}, y: {1}, vec: {2}'.format(self.x, self.y, self.vector)
    
    def move(self):
        if (self.y == 14):
            self.x = int(4)
            self.y = int(0)
            self.vector = [1, 1]
        elif (self.x == 0):
            self.vector[0] = 1            
            self.x += self.vector[0]
            self.y += self.vector[1] 
        elif (self.x == 8):
            self.vector[0] = -1 
            self.x += self.vector[0]
            self.y += self.vector[1] 
        else:    
            self.x += self.vector[0]
            self.y += self.vector[1]        

class Bullet(object):
    def __init__(self, x:int):
        self.x = int(x)
        self.y = int(14)
        self.mark = 'e'
    def __repr__(self):
        return 'x: {0}, y: {1}'.format(self.x, self.y)
    def move(self):
        self.y -= 1  

class Gun(object):
    def __init__(self):
        self.x = int(4)
        self.mark = 'X'
        # 残弾数
        self.remain_bullet = 2
    def __repr__(self):
        return 'x: {0}, 残弾数: {1}'.format(self.x, self.remain_bullet)

class Game(object):
    def __init__(self):
        self.board = [
            ['|', '-', '-', '-', 'V', '-', '-', '-', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', ' ', ' ', ' ', ' ', ' ', ' ', ' ', '|'],
            ['|', '-', '-', '-', 'X', '-', '-', '-', '|'],            
        ]
        self.enemy = Enemy()
        self.bullets = []
        self.gun = Gun()
        self.point = 0
        self.fail = 0
        # 初期段階
        self.phase = 0
        
    def print_board(self):
        row = len(self.board)        
        print()
        col = len(self.board[0])
        for i in range(0, row):
            row_str = ''
            for j in range(0, col):
                row_str += self.board[i][j]
            print(row_str) 
            
    def move_gun(self, key):
        if (key == 'j'):
            if (self.gun.x > 0):
                # 元の位置を更新
                self.board[14][self.gun.x] = ' '
                self.gun.x -= 1
                if (self.gun.x < 0):
                    self.gun.x = 0
                # 移動後の位置を更新
                self.board[14][self.gun.x] = 'X'
        elif (key == 'l'):
            if (self.gun.x < 8):
                # 元の位置を更新
                self.board[14][self.gun.x] = ' '
                self.gun.x += 1
                if (self.gun.x > 8):
                    self.gun.x = 8
                # 移動後の位置を更新
                self.board[14][self.gun.x] = 'X'       
    
    def fire_gun(self, key):
        if (self.gun.remain_bullet > 0 and key == 'i'):
            self.bullets.append(Bullet(self.gun.x))
            self.gun.remain_bullet -= 1           
    
    def update_enemy_board(self):
        # 元の位置を更新
        if (self.enemy.x == 4 and self.enemy.y == 0):
            self.board[0][4] = 'V'
        elif (self.enemy.x == self.gun.x and self.enemy.y == 14):
            self.board[14][self.gun.x] = 'X'
        elif (self.enemy.x == 0 or self.enemy.x == 8):
            self.board[self.enemy.y][self.enemy.x] = '|'
        else:
            self.board[self.enemy.y][self.enemy.x] = ' '
        # fail
        if (self.enemy.y == 14):
            self.fail += 1
            # reset gun bullet
            self.gun.remain_bullet = 2
        # 移動
        self.enemy.move()
        # 移動後を更新
        self.board[self.enemy.y][self.enemy.x] = 'O'
    
    def update_bullets_board(self):
        for index, bullet in enumerate(self.bullets):
            if (bullet.y >= 0):
                # 元の位置のマークが次のenemyの位置かどうか
                is_enemy = (self.board[bullet.y][bullet.x] == 'O')
                if (not is_enemy and bullet.y < 14):
                    if (bullet.x == 4 and bullet.y == 0):
                        self.board[0][4] = 'V'
                    elif (bullet.x == self.gun.x and bullet.y == 14):
                        self.board[14][self.gun.x] = 'X'
                    elif (bullet.x == 0 or bullet.x == 8):
                        self.board[bullet.y][bullet.x] = '|'
                    else:
                        self.board[bullet.y][bullet.x] = ' '
                # 移動
                bullet.move()
                # 移動後を更新
                if (bullet.y >= 0):
                    # 移動後がenemyと当たった場合
                    if (self.board[bullet.y][bullet.x] == 'O'):                        
                        if (bullet.x == 4 and bullet.y == 0):
                            self.board[0][4] = 'V'
                        elif (bullet.x == self.gun.x and bullet.y == 14):
                            self.board[14][self.gun.x] = 'X'
                        elif (bullet.x == 0 or bullet.x == 8):
                            self.board[bullet.y][self.bullet.x] = '|'
                        else:
                            self.board[bullet.y][bullet.x] = ' '
                        bullet.y = -1
                        self.point += 1
                        self.enemy = Enemy()
                        self.gun.remain_bullet = 2    
                    else:    
                        self.board[bullet.y][bullet.x] = 'e'  
                    
    def update_phase(self, key, rand_num):
        if (key == 'i' or key == 'j' or key == 'k' or key == 'l'):            
            self.phase += 1
            self.gun.x = rand_num        
            self.move_gun(key)
            self.fire_gun(key)
            self.update_enemy_board()
            self.update_bullets_board()
```

```python
game = Game()
print('phase: {0}, point: {1}, fail: {2}, 残弾数{3}'.format(game.phase, game.point, game.fail, game.gun.remain_bullet))
game.print_board()

while (game.fail < 5):    
    key = 'i'
    rand_num = random.randrange(9)
    tmp = rand_num % 4
    if (rand_num == 0):
        key = 'i'
    elif (rand_num == 3):
        key = 'j'
    elif (rand_num == 2):
        key == 'k'
    else:
        key == 'l'
    game.update_phase(key, rand_num)
    print('phase: {0}, point: {1}, fail: {2}, 残弾数{3}'.format(game.phase, game.point, game.fail, game.gun.remain_bullet))
    game.print_board()
    print('=' * 20)
```
